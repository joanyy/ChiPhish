```python
def show_re(js_data1,js_data2,js_data3,col,val2):
  # print('js_data1 is',js_data1)
  print('col1 is %s, col2 is %s'%(col,val2))
  col1_ave=round(js_data1[col].mean(),2)
  col1_std=round(js_data1[col].std(),3)
  val1_ave=round(js_data1[val2].mean(),2)
  val1_std=round(js_data1[val2].std(),3)

  # print('js_data2 is',js_data2)
  print('col1 is %s,col2 is %s'%(col,val2))
  col2_ave=round(js_data2[col].mean(),2)
  col2_std=round(js_data2[col].std(),2)
  val2_ave=round(js_data2[val2].mean(),2)
  val2_std=round(js_data2[val2].std(),3)

  # print('js_data3 is',js_data3)
  print('col1 is %s,col2 is %s'%(col,val2))
  col3_ave=round(js_data3[col].mean(),2)
  col3_std=round(js_data3[col].std(),3)
  val3_ave=round(js_data3[val2].mean(),2)
  val3_std=round(js_data3[val2].std(),3)


  print("\\res{%.2f}{%.3f}& \\res{%.2f}{%.3f} & \\res{%.2f}{%.3f} & \\res{%.2f}{%.3f} & \\res{%.2f}{%.3f} & \\res{%.2f}{%.3f}"%(col1_ave,col1_std,val1_ave,val1_std,col2_ave,col2_std,val2_ave,val2_std,col3_ave,col3_std,val3_ave,val3_std))
```
