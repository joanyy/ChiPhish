```python
train_en,test_en=train_test_split(data_en,test_size=0.2,shuffle=True)
 
train_en_x=train_en.iloc[:,0:65]
train_en_y=train_en.iloc[:,65] 
test_en_x=test_en.iloc[:,0:65] 
test_en_y=test_en.iloc[:,65] 

scaler_entire_en=preprocessing.MinMaxScaler()
scaler_entire_en.fit(train_en_x)
train_x_en=scaler_entire_en.transform(train_en_x)
test_x_en=scaler_entire_en.transform(test_en_x)

#use only html features
scaler_html_en=preprocessing.MinMaxScaler()
scaler_html_en.fit(train_en.iloc[:,36:65])
train_html_x_en=scaler_html_en.transform(train_en.iloc[:,36:65])
test_html_x_en=scaler_html_en.transform(test_en.iloc[:,36:65])
#print('train_en html',train_en.columns[25:49])
#only url features
scaler_url_en=preprocessing.MinMaxScaler()
scaler_url_en.fit(train_en.iloc[:,0:36])
#print('train ch url is',train_en.columns[0:25])

train_url_x_en=scaler_url_en.transform(train_en.iloc[:,0:36])
test_url_x_en=scaler_url_en.transform(test_en.iloc[:,0:36])

test_url_y_en=test_html_y_en=test_en_y
train_url_y_en=train_html_y_en=train_en_y
print('combine_rf_whole:')# train_en_x

rf_start=time.time()
model_rf1_en=RandomForestClassifier(criterion='gini', max_depth=45, min_samples_split=3, n_estimators=200)
model_rf_com,rf_tr_time,rf_te_time,rf_tnr_com,rf_recall_com,rf_f1_com=time_metri(model_rf1_en,train_x_en,train_en_y,test_x_en,test_en_y)#using normalized data to train, so all the model_rf4_com data must normalized

print('combine_lr_whole:')
model_lr1_en = LogisticRegression(max_iter=500,penalty='l2',C=100)
model_lr_com,lr_tr_time,lr_te_time,lr_tnr_com,lr_recall_com,lr_f1_com =time_metri(model_lr1_en,train_x_en,train_en_y,test_x_en,test_en_y)

print('dt=======')
model_dt1_en=DecisionTreeClassifier(max_depth=38, min_samples_leaf= 1, min_samples_split=2)
model_dt_com,dt_tr_time,dt_te_time,dt_tnr_com,dt_recall_com,dt_f1_com =time_metri(model_dt1_en,train_x_en,train_en_y,test_x_en,test_en_y)

print('knn======')
model_knn1_en = KNeighborsClassifier(n_neighbors=3)
model_knn_com,knn_tr_time,knn_te_time,knn_tnr_com,knn_recall_com,knn_f1_com= time_metri(model_knn1_en,train_x_en,train_en_y,test_x_en,test_en_y)

print('mlp=====')
model_mlp1_en=MLPClassifier(activation='relu', max_iter=1000, solver='adam',learning_rate='adaptive')
model_mlp_com,mlp_tr_time,mlp_te_time,mlp_tnr_com,mlp_recall_com,mlp_f1_com=time_metri(model_mlp1_en,train_x_en,train_en_y,test_x_en,test_en_y)

print('ab======')
model_ada_en=AdaBoostClassifier(learning_rate=1.5,n_estimators=300)
model_ada_com,ab_tr_time,ab_te_time,ab_tnr_com,ab_recall_com,ab_f1_com=time_metri(model_ada_en,train_x_en,train_en_y,test_x_en,test_en_y)

print('svm=======')

model_svm_en=svm.SVC(C= 500, gamma= 1, kernel= 'linear')
model_svm_com,svm_tr_time,svm_te_time,svm_tnr_com,svm_recall_com,svm_f1_com=time_metri(model_svm_en,train_x_en,train_en_y,test_x_en,test_en_y)
print('gb========')
model_gb_en=GradientBoostingClassifier(learning_rate= 0.2, n_estimators= 300)

model_gb_com,gb_tr_time,gb_te_time,gb_tnr_com,gb_recall_com,gb_f1_com=time_metri(model_gb_en,train_x_en,train_en_y,test_x_en,test_en_y)
# for cnn models
one_hot_train_y_en=tf.one_hot(indices=train_en_y.array,depth=2,on_value=1.0,off_value=0.0)
one_hot_test_y_en=tf.one_hot(indices=test_en_y.array,depth=2,on_value=1.0,off_value=0.0)
test_x_re_en=np.asarray(test_x_en).reshape(test_x_en.shape[0],test_x_en.shape[1],1)
train_x_re_en=np.asarray(train_x_en).reshape(train_x_en.shape[0],train_x_en.shape[1],1)
train_html_x_re_en=np.asarray(train_html_x_en).reshape(train_html_x_en.shape[0],train_html_x_en.shape[1],1)
test_html_x_re_en=np.asarray(test_html_x_en).reshape(test_html_x_en.shape[0],test_html_x_en.shape[1],1)
train_url_x_re_en=np.asarray(train_url_x_en).reshape(train_url_x_en.shape[0],train_url_x_en.shape[1],1)
test_url_x_re_en=np.asarray(test_url_x_en).reshape(test_url_x_en.shape[0],test_url_x_en.shape[1],1)

print('cnn==========')
model_cnn_combined=create_model_cnn_full()
cnn_start=time.time()
model_cnn_combined.fit(train_x_re_en,one_hot_train_y_en,batch_size=32,epochs=50, verbose=2)

cnn_end=time.time()
cnn_tr_time=cnn_end-cnn_start

predict_y_com=model_cnn_combined.predict(test_x_re_en)
pre_class_com=np.argmax(predict_y_com,axis=1)
confusion_mtx_com = confusion_matrix(test_en_y, pre_class_com) 
cnn_tnr_com,cnn_recall_com,cnn_f1_com=get_fpr(test_en_y,pre_class_com)

cnn_te_time=time.time()-cnn_end 

print('tpr is %.2f, tnr is %.2f, f1 is %.2f'%(cnn_recall_com,cnn_tnr_com,cnn_f1_com))
print('cnn_tr_time is',cnn_tr_time)
print('cnn_te_time is',cnn_te_time)
```
