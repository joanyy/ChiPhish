# performance of custom PWD training on ChiPhish

```python
ch_feature='../data/chiphish/chspec_combine_chphish_full.json' 
ch_data=load_dataset(ch_feature)
  
ch_data['label'].value_counts() 
data_ch=ch_data.iloc[:,:66]
data_ch=data_ch.astype(float)
print('data_ch',data_ch.shape)
```

# Custom Fc-based PWD training on ChiPhish ,
1. f1 value refer to the red box on left side of Fig.8c,
2. tpr and tnr refer to Table VIII, bottom cell of the right column,
3. te_time and tr_time refer to the right column of Table VII

```python
train_ch,test_ch=train_test_split(data_ch,test_size=0.2,shuffle=True) 

train_ch_x=train_ch.iloc[:,0:65]
train_ch_y=train_ch.iloc[:,65] 
test_ch_x=test_ch.iloc[:,0:65] 
test_ch_y=test_ch.iloc[:,65] 

scaler_entire_ch=preprocessing.MinMaxScaler()
scaler_entire_ch.fit(train_ch_x)
train_x_ch=scaler_entire_ch.transform(train_ch_x)
test_x_ch=scaler_entire_ch.transform(test_ch_x)

#use only html features
scaler_html_ch=preprocessing.MinMaxScaler()
scaler_html_ch.fit(train_ch.iloc[:,36:65])
train_html_x_ch=scaler_html_ch.transform(train_ch.iloc[:,36:65])
test_html_x_ch=scaler_html_ch.transform(test_ch.iloc[:,36:65])
#print('train_ch html',train_ch.columns[25:49])
#only url features
scaler_url_ch=preprocessing.MinMaxScaler()
scaler_url_ch.fit(train_ch.iloc[:,0:36])
#print('train ch url is',train_ch.columns[0:25])

train_url_x_ch=scaler_url_ch.transform(train_ch.iloc[:,0:36])
test_url_x_ch=scaler_url_ch.transform(test_ch.iloc[:,0:36])

test_url_y_ch=test_html_y_ch=test_ch_y
train_url_y_ch=train_html_y_ch=train_ch_y

print('combine_rf_whole:')


rf_start=time.time()
model_rf1_ch=RandomForestClassifier(criterion='gini', max_depth=25, min_samples_split=3, n_estimators=100)
model_rf_com,rf_tr_time,rf_te_time,rf_tnr_com,rf_recall_com,rf_f1_com=time_metri(model_rf1_ch,train_x_ch,train_ch_y,test_x_ch,test_ch_y)#using normalized data to train, so all the model_rf4_com data must normalized
print('combine_lr_whole:')
model_lr1_ch = LogisticRegression(max_iter=500,penalty='l2',C=10) 
model_lr_com,lr_tr_time,lr_te_time,lr_tnr_com,lr_recall_com,lr_f1_com =time_metri(model_lr1_ch,train_x_ch,train_ch_y,test_x_ch,test_ch_y)

print('dt=====')
model_dt1_ch=DecisionTreeClassifier(max_depth=16, min_samples_leaf= 1, min_samples_split=2) 
model_dt_com,dt_tr_time,dt_te_time,dt_tnr_com,dt_recall_com,dt_f1_com  =time_metri(model_dt1_ch,train_x_ch,train_ch_y,test_x_ch,test_ch_y)

print('knn======')
model_knn1_ch = KNeighborsClassifier(n_neighbors=3)
model_knn_com,knn_tr_time,knn_te_time,knn_tnr_com,knn_recall_com,knn_f1_com= time_metri(model_knn1_ch,train_x_ch,train_ch_y,test_x_ch,test_ch_y)

print('mlp=======')
model_mlp1_ch=MLPClassifier(activation='relu', max_iter=2500, solver='adam',learning_rate='adaptive')
model_mlp_com,mlp_tr_time,mlp_te_time,mlp_tnr_com,mlp_recall_com,mlp_f1_com=time_metri(model_mlp1_ch,train_x_ch,train_ch_y,test_x_ch,test_ch_y)

print('ab======')
model_ada_ch=AdaBoostClassifier(learning_rate=0.8,n_estimators=100)
model_ada_com,ab_tr_time,ab_te_time,ab_tnr_com,ab_recall_com,ab_f1_com=time_metri(model_ada_ch,train_x_ch,train_ch_y,test_x_ch,test_ch_y)

print('svm======')
model_svm_ch=svm.SVC(C= 100, gamma=1, kernel="linear")
model_svm_com,svm_tr_time,svm_te_time,svm_tnr_com,svm_recall_com,svm_f1_com=time_metri(model_svm_ch,train_x_ch,train_ch_y,test_x_ch,test_ch_y)

print('gb========')
model_gb_ch=GradientBoostingClassifier(learning_rate= 0.15, n_estimators=100)
model_gb_com,gb_tr_time,gb_te_time,gb_tnr_com,gb_recall_com,gb_f1_com=time_metri(model_gb_ch,train_x_ch,train_ch_y,test_x_ch,test_ch_y)

one_hot_train_y_ch=tf.one_hot(indices=train_ch_y.array,depth=2,on_value=1.0,off_value=0.0)
one_hot_test_y_ch=tf.one_hot(indices=test_ch_y.array,depth=2,on_value=1.0,off_value=0.0)
test_x_re_ch=np.asarray(test_x_ch).reshape(test_x_ch.shape[0],test_x_ch.shape[1],1)
train_x_re_ch=np.asarray(train_x_ch).reshape(train_x_ch.shape[0],train_x_ch.shape[1],1)
train_html_x_re_ch=np.asarray(train_html_x_ch).reshape(train_html_x_ch.shape[0],train_html_x_ch.shape[1],1)
test_html_x_re_ch=np.asarray(test_html_x_ch).reshape(test_html_x_ch.shape[0],test_html_x_ch.shape[1],1)
train_url_x_re_ch=np.asarray(train_url_x_ch).reshape(train_url_x_ch.shape[0],train_url_x_ch.shape[1],1)
test_url_x_re_ch=np.asarray(test_url_x_ch).reshape(test_url_x_ch.shape[0],test_url_x_ch.shape[1],1)

print('cnn========')
model_cnn_combined=create_model_cnn_full()
cnn_start=time.time()
model_cnn_combined.fit(train_x_re_ch,one_hot_train_y_ch,batch_size=32,epochs=50, verbose=2)
cnn_end=time.time()
cnn_tr_time=cnn_end-cnn_start

predict_y_com=model_cnn_combined.predict(test_x_re_ch)
pre_class_com=np.argmax(predict_y_com,axis=1)
confusion_mtx_com = confusion_matrix(test_ch_y, pre_class_com) 
cnn_tnr_com,cnn_recall_com,cnn_f1_com=get_fpr(test_ch_y,pre_class_com)
cnn_te_time=time.time()-cnn_end

print('tpr is %.2f, tnr is %.2f, f1 is %.2f'%(cnn_recall_com,cnn_tnr_com,cnn_f1_com))
print('cnn_tr_time is',cnn_tr_time)
print('cnn_te_time is',cnn_te_time)
```
